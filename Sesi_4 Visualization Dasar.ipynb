Basic Visualization

```python
import numpy as np
import pandas as pd
```

```python
df_can = pd.read_excel('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Canada.xlsx',
                      sheet_name='Canada by Citizenship',
                      skiprows=range(20),
                      skipfooter=2)

print('Data read into a pandas dataframe')
```

```python
df_can.head()
```

```python
df_can.tail()
```

```python
df_can.info()
```

```python
df_can.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'],
           axis=1, inplace=True)

df_can.head(2)
```

```python
df_can.rename(columns={
    'OdName': 'Country',
    'AreaName': 'Continent',
    'RegName': 'Region'
}, inplace=True)

df_can.head(2)
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

```python
df_can.head(2)
```

```python
df_can[['Country', 2010, 2011, 2012, 2013]]
```

```python
df_can.set_index('Country', inplace=True)
```

```python
df_can.head(3)
```

```python
df_can.loc['Indonesia']
```

```python
df_can.columns = list(map(str, df_can.columns))
```

```python
years = list(map(str, range(1980, 2014)))

years
```

```python
df_can[
    df_can['Continent'] == 'Asia'
]
```

```python
df_can[
    df_can['Region'] == 'South-Eastern Asia'
]
```

## Visualizing Data using Matplotlib

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print(plt.style.available)
```

```python
mpl.style.use(['ggplot'])
```

## Line Plots

```python
haiti = df_can.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int)
haiti.plot()

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```

```python
haiti.plot()

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.text(2000, 6000, '2010 Earthquake')

plt.show()
```

```python
China_India = df_can.loc[['China', 'India'], years]

China_India
```

```python
China_India.plot(kind='line')
```

```python
df_CI = China_India.transpose()

df_CI.head()
```

```python
df_CI.plot()

plt.title('Immigrants from China and India')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
```

```python
df_top5 = df_can.head(5)
```

```python
df_top5 = df_top5[years].transpose()
df_top5.head()
```

```python
df_top5.plot(figsize=(14, 8))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

## Area Plot

```python
df_top5.plot(kind='area')
```

```python
df_top5.plot(kind='area',
            stacked=False,
            alpha=0.25,
            figsize=(20, 10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

## Bar Chart

```python
df_iceland = df_can.loc['Iceland', years]
```

```python
df_iceland.plot(kind='bar',
               figsize=(10, 6))

plt.title('Icelandic Immigrants to Canada')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

```python
df_top15 = df_can['Total'].head(15)
```

```python
df_top15.plot(kind='barh',
             figsize=(10, 10),
             color='plum')
plt.title('Top 15 Countries contributing to immigration to Canada')

plt.show()
```

## Pie Chart

```python
df_continents = df_can.groupby('Continent', axis=0).sum()
```

```python
df_continents
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize=(15, 6),
                           autopct='%1.1f%%', # 1.1f 1angka dibelakang koma dalam bentuk f/float
                           startangle=90,
                           shadow=True)

plt.title('Immigration to Canada by Continent')
plt.axis('equal')

plt.show()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral',
              'lightskyblue', 'lightgreen', 'plum']

explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents['Total'].plot(kind='pie',
                           figsize=(15, 6),
                           autopct='%1.1f%%', # 1.1f 1angka dibelakang koma dalam bentuk f/float
                           startangle=90,
                           shadow=True,
                           labels=None,
                           pctdistance=1.12,
                           colors=colors_list,
                           explode=explode_list)

plt.title('Immigration to Canada by Continent', y=1.12)
plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

## Box Plot

```python
df_japan = df_can.loc[['Japan'], years].transpose()
```

```python
df_japan.plot(kind='box',
             figsize=(8, 6))

plt.title('Box Plot of Japanese Immigrants to Canada')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
df_CI.plot(kind='box',
          figsize=(8, 6),
          vert=False)

plt.title('Box Plot of China and India Immigrants to Canada')
plt.ylabel('Number of Immigrants')

plt.show()
```

## Subplots

```python
fig = plt.figure()

ax0 = fig.add_subplot(1, 2, 1)
ax1 = fig.add_subplot(1, 2, 2)

df_CI.plot(kind='box',
          figsize=(18, 6),
          vert=False,
          ax=ax0)
ax0.set_title('Box plot of Immigrants from India and China')
ax0.set_xlabel('Number of Immigrants')
ax0.set_ylabel('Countries')

df_CI.plot(kind='line',
          figsize=(18, 6),
          ax=ax1)
ax1.set_title('Line plot of Immigrants from India and China')
ax1.set_xlabel('Number of Immigrants')
ax1.set_ylabel('Years')

plt.show()
```

## Scatter Plots

```python
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

df_tot.head()
```

```python
df_tot.index = map(int, df_tot.index)
df_tot.reset_index(inplace=True)
df_tot.head()
```

```python
df_tot.columns = ['years', 'total']
df_tot.head()
```

```python
df_tot.plot(kind='scatter',
           x='years',
           y='total',
           figsize=(10, 6))
plt.title('Total Immigration to Canada')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```
